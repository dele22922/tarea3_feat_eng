```python
## Importamos las librerias
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Definir los arcos y su respectiva capacidad

arcos = [(1,2,6),(1,3,4),(2,4,3),(2,5,2),(3,4,2),(3,5,5),(4,6,6),(5,6,4)]
```

```python
## Crear el grafo vacio y luego agregarle los arcos.  Cada arco debe tener un parametro llamado "capacity" que define la
## capacidad de dicho arco.

redflujo = nx.DiGraph()
for x in arcos:
    redflujo.add_edge(x[0],x[1], capacity=x[2])
```

```python
## Visualizar el grafo
plt.figure(1,figsize=(6,6))
nx.draw(redflujo, with_labels=True)
plt.show()
```

```python
##algoritmo faulfolkerson
## Calculamos el flujo maximo, es decir, lo mas que puede fluir por la red entre 2 nodos, en este caso, el 1 y el 6

nx.maximum_flow(redflujo,1,6)
```
