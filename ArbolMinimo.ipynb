## Modelo del Arbol Mínimo de Expansión

### Este modelo busca encontrar la forma más barata de conectar todos los nodos de una red, utilizando el peso (distancia) de los arcos.


```python
import networkx as nx
```

La ciudad de Vancouver está plani
cando el desarrollo de una nueva linea en sistemas de tránsito. El sistema debe conectar 8 puntos distintos entre residencias, centros comerciales y otros puntos importantes. El departamento de planificación de tránsito de la ciudad necesita seleccionar un conjunto de rutas que conecten los distintos puntos al menor costo posible. La red seleccionada debe considerar la factibilidad y el costo de cada ruta.

```python
## En este caso, vamos a crear el grafo a partir de una lista de arcos, donde cada arco está dado por un tuple que
## contiene 3 elementos, el nodo de inicio, el nodo final y la distancia

arcos = [(1,3,33),(1,2,28),(2,3,34),(3,4,30),(2,4,32),(1,7,40),(2,7,37),(3,5,50),(2,5,35),(4,5,39),(2,6,41),(6,7,36),
          (5,6,45),(6,8,43),(5,8,38),(3,8,55),(7,8,44)]
```

```python
grafo1 = nx.Graph()
for x in arcos:
    grafo1.add_edge(x[0],x[1], weight=x[2])
```

```python
nx.draw_networkx(grafo1, with_labels=True)
```

```python
arbol = nx.minimum_spanning_tree(grafo1)
```

```python
nx.draw_networkx(arbol, with_labels=True)
```

```python
arcosarbol = nx.minimum_spanning_edges(grafo1)
```

```python
listarcos = []
for x in arcosarbol:
    listarcos.append(x)
```
