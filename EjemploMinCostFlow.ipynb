```python
## Primero cargamos la libreria para redes

import networkx as nx
```

```python
## Creamos una lista de nodos y una lista con las demandas (demanda negativa es igual a oferta)

listanodos = ['P1','P2','CD N','CD','CD S','L1','L2']

demandas = [-1000, -1500, 0, 0, 0, 1250, 1250]
#oferta se representa como una demanda negativa
```

```python
## Creamos el grafo vacio y luego agregamos los nodos con sus respectivas demandas, este grafo es dirigido
grafo = nx.DiGraph()

for i,x in enumerate(listanodos):
    grafo.add_node(x, demand=demandas[i])
```

```python
grafo.nodes
```

```python
## Definimos los arcos por medio de listas, indicando el inicio, el fin, el peso (costo) y la capacidad de cada arco


inicio = ['P1','P1','P2','P2','CD N','CD N','CD','CD','CD S','CD S']
fin = ['CD N','CD S','CD N','CD S','L1','CD','L1','L2','CD','L2']
costo = [1,2,2,3,6,1,3,4,2,5]
capacidad = [900]*10
```

```python
capacidad
```

```python
## Agregar los arcos al graf

for i in range(len(inicio)):
    grafo.add_edge(inicio[i],fin[i],weight=costo[i],capacity=capacidad[i])
```

```python
## Dibujar el grafo definiendo las coordenadas de los nodos

coord = [(1,3.5),(1,2),(3,5),(4,3),(3,1),(6,3.5),(6,2)]

posiciones = dict(zip(listanodos,coord))

#nx.draw(grafo, with_labels=True, pos=posiciones)
nx.draw_networkx(grafo, with_labels=True, pos=posiciones)
```

```python
print(nx.min_cost_flow(grafo)) ## generar el flujo de costo minimo
```

```python
print(nx.min_cost_flow_cost(grafo))  ## calcular el costo minimo
```
