```python
arcos_a
```

```python
grafo1 = nx.Graph()
for x in arcos_a:
    grafo1.add_edge(x[0],x[1], weight=x[2])
```

```python
nx.draw_networkx(grafo1, with_labels=True)
```

```python
arbol = nx.minimum_spanning_tree(grafo1)
```

```python
nx.draw_networkx(arbol, with_labels=True)
```

```python
arcosarbol2 = nx.minimum_spanning_edges(grafo1)
```

```python
listarcos2 = []
for x in arcosarbol2:
    listarcos2.append(x)
```
