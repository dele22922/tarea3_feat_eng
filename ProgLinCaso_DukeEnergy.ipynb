## Programacion Lineal Entera Mixta

Usted ha sido designado por el gerente de su empresa para decidir cmo distribuir su trfico telefnico en el prximo mes, seleccionando entre 3 proveedores posibles y asignando la cantidad de trfico (minutos) que desee en cada caso, es decir, puede repartir el trfico en 1, 2 o 3 proveedores y su decisin slo depender de los costos de cada alternativa.

El proveedor 1 cobra un cargo fijo mensual de 150 y el costo por minuto a red fija es de  0.02 y a celular de 0.12. El proveedor 2 tiene un cargo fijo mensual de 160, con un costo por minuto de 0.015 y 0.15 a red fija y celular respectivamente. Finalmente el proveedor 3 tiene un cargo fijo mensual de solamente 50 con un costo por minuto a red fija de 0.03 y a celular de 0.14. Si usted llama por uno de estos proveedores (aunque hable slo un minuto) deber pagar el cargo fijo. Asuma que la cantidad de minutos que la empresa consume mensualmente es de 30,000 para red fija y 18,000 para celular. Formule y resuelva un modelo de Programacin que permita decidir cmo distribuir el trfico telefnico mensual de la forma ms econmica para la empresa.


```python
## Datos

reqRedFija = 30000
reqCelular = 18000
CostosFijos = np.array([150,160,50])
CargosMinFijo = np.array([0.02,0.015,0.03])
CargosMinCel = np.array([0.12,0.15,0.14])
numprov = len(CostosFijos)
```

```python
## Problema

telefonia = LpProblem("Telefonia", LpMinimize)
```

```python
## Variables de Decision

mincelular =  LpVariable.matrix("Cel", range(numprov),lowBound=0,cat="Continuous")
minfija = LpVariable.matrix("Fija", range(numprov), lowBound=0, cat="Continuous")
usarprov = LpVariable.matrix("Prov", range(numprov), cat="Binary")
```

```python
## Funcion objetivo

CostoMinFijo =  lpSum([minfija[i]*CargosMinFijo[i] for i in range(numprov)])
CostoMinCel = lpSum([mincelular[i]*CargosMinCel[i] for i in range(numprov)])
CostoFijoTot = lpSum([CostosFijos[i]*usarprov[i] for i in range(numprov)])

telefonia += CostoMinFijo + CostoMinCel + CostoFijoTot
```

```python
## Restricciones

telefonia +=  lpSum([mincelular[i] for i in range(numprov)]) >= reqCelular
telefonia +=  lpSum([minfija[i] for i in range(numprov)]) >= reqRedFija

for j in range(numprov):
    telefonia +=  mincelular[j] + minfija[j] <= (reqRedFija + reqCelular) * usarprov[j]
```

```python
telefonia
```

```python
telefonia.solve()
```

```python
telefonia.objective.value()
```

```python
for var in telefonia.variables():
    print(var.name, var.value())
```
